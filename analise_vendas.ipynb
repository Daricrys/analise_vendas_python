```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar gráficos para aparecerem no notebook
%matplotlib inline
```

```python
df = pd.DataFrame({
    "Produto": ["Notebook", "Notebook", "Mouse", "Teclado", "Mouse", "Monitor", "Teclado", "Notebook", "Monitor", "Mouse"],
    "Vendas": [3500, 4200, 150, 200, 120, 900, 250, 4000, 1000, 180],
    "Quantidade": [3, 4, 10, 5, 8, 2, 6, 3, 1, 12]
})

df
```

```python
# Estatísticas descritivas de vendas
print("Estatísticas de Vendas:")
print(df["Vendas"].describe())

# Quantidade total por produto
print("\nQuantidade total por produto:")
print(df.groupby("Produto")["Quantidade"].sum())
```

```python
sns.barplot(x="Produto", y="Vendas", data=df, estimator=sum)
plt.title("Vendas Totais por Produto")
plt.show()
```

```python
sns.scatterplot(x="Quantidade", y="Vendas", data=df, hue="Produto", s=100)
plt.title("Relação entre Quantidade e Vendas")
plt.show()
```
